```python
#Pandas is a library .It handle table ,file reading,database related one.
#Import means load all function from particular library
#as used for short form
import pandas as pd
```

```python
#Create variable for load the dataset
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#print the dataset
dataset
```

```python
#assign independent variable we take years experience as input in this table
#we can use this method to specify particular column in this table
independent=dataset[["YearsExperience"]]
```

```python
#print input variable
independent
```

```python
#Dependent variable we specify salary as output in this table
dependent=dataset[["Salary"]]
dependent
```

```python
#sklearn library used to split train and test model
from sklearn.model_selection import train_test_split
#Then call train_test_split function it is parameterised function
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#This is learning phase.
#using linear regression function we create the model
from sklearn.linear_model import LinearRegression
#create variable for linear regression
regressor =LinearRegression()
#we send the training set input and training set output.after the execution of fit method weight and bias are calculated
regressor.fit(X_train,Y_train)
```

```python
#To calculate weight
weight=regressor.coef_
weight
```

```python
#To calculate bias
bias=regressor.intercept_
bias
```

```python
#Evalutionmetric test is for crosscheck the learning of AI.predict is the function of application.
#Evalutionmetric test is done for test data input sowe give X_test
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
#to find r2_score,we take actual output and predictor.
r_score=r2_score(Y_test,y_pred)
```

```python
#r2_score is 1-better performance
#r2_score is 0-poor performance
r_score
```

```python
#pickle is used to save model creation
import pickle
#Create filename,it is pickle extension so we save .sav
filename="finalized_model_linear.sav"
```

```python
#open the file in folder and save the model ,it helpful for end user
pickle.dump(regressor, open(filename,"wb"))
```

```python
#load the model and rb is used for just read 
loaded_model=pickle.load(open("finalized_model_linear.sav","rb"))
#we check it and for prediction we can do it
result=loaded_model.predict([[15]])
```
